```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
def scraper(url):
    
    from selenium import webdriver
    from urllib.parse import urljoin
    import time


    headers = {
        'authority': 'scrapeme.live',
        'dnt': '1',
        'upgrade-insecure-requests': '1',
        'user-agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.61 Safari/537.36',
        'accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,/;q=0.8,application/signed-exchange;v=b3;q=0.9',
        'sec-fetch-site': 'none',
        'sec-fetch-mode': 'navigate',
        'sec-fetch-user': '?1',
        'sec-fetch-dest': 'document',
        'accept-language': 'en-GB,en-US;q=0.9,en;q=0.8',
    }

    driver = webdriver.Chrome(
        executable_path=r"/home/paritosh/PycharmProjects/zomato/chromedriver")  # this might need a change since i am working in pycharm you might need to import os

    driver.get(url)  # url of the website of particular cities you want to scrap.

      # Allow 2 seconds for the web page to (open depends on you)
    scroll_pause_time = 0  # You can set your own pause time. dont slow too slow that might not able to load more data
    screen_height = driver.execute_script("return window.screen.height;")  # get the screen height of the web
    i = 1

    while True:
        # scroll one screen height each time
        driver.execute_script("window.scrollTo(0, {screen_height}*{i});".format(screen_height=screen_height, i=i))
        i += 1
        time.sleep(scroll_pause_time)
        # update scroll height each time after scrolled, as the scroll height can change after we scrolled the page
        scroll_height = driver.execute_script("return document.body.scrollHeight;")
        # Break the loop when the height we need to scroll to is larger than the total scroll height
        time.sleep(0.5)
        if (screen_height) * i > scroll_height:
            break
    
    return driver.page_source
```

```python
final_lst=[]
for p in range(1,21):
    url_scp="https://www.amazon.in/s?k=bags&crid=2M096C61O4MLT&qid=1653308124&sprefix=ba%2Caps%2C283&ref=sr_pg_"+str(p)

    soup = BeautifulSoup(scraper(url_scp), "html.parser")
    x=soup.find_all("div", class_="a-section a-spacing-small a-spacing-top-small")
    count=0
    for i in x:
        try:
            item_name=i.find_all(class_="a-size-mini a-spacing-none a-color-base s-line-clamp-2")[0].text
            
            item_price=float((i.find_all(class_="a-price-whole")[0].text).replace(',',''))
            
            rating= float((i.find_all("div",class_="a-section a-spacing-none a-spacing-top-micro")[0].text).split()[0])
            
            rev = float((i.find_all(class_="a-size-base s-underline-text")[0].text).replace(',',''))
            
            try:
                url="https://www.amazon.in"+(i.findAll(class_="sg-row")[0].find_all("a")[1])['href']
                
            except:
                url=np.nan  
                
            final_lst.append([item_name,item_price,rating,rev,url])

        except:
            pass
    print("##########################page"+str(p)+"############################")
```

```python
df = pd.DataFrame(final_lst, columns = ['Product_name','Product_price','Product_rating','No_of_reviews','Product_url'])
```

```python
#Removing missing urls
new_df=new_df.dropna()
```

```python
#Scraping data from url

final_lst=[]
count=0
for data in new_df.values:
    count=count+1
    print(count)
    link=data[-1]
    soup = BeautifulSoup(scraper(link), "html.parser")

    #type check
    x=soup.find_all("ul",class_="a-unordered-list a-nostyle a-vertical a-spacing-none detail-bullet-list")

    if len(x) != 0:
        print("type 1")
        x1=soup.find_all(class_="a-list-item")

        #ASIN
        for i in x1:
            if "ASIN" in i.text:
                ASIN=i.text.split()[-1]

        # Manufacturer
        for i in x1:
            if "Manufacturer" in i.text:
                 Manufacturer = i .text.replace(" ","").split()[-1]
        #Desc
        desc = soup.find_all("div",class_="a-section a-spacing-medium a-spacing-top-small")
        desc=desc[0].text
        final_lst.append([data[0],data[1],data[2],data[3],data[4],ASIN,Manufacturer,desc])
    else:
        print("type 2")
        lst2=[]
        lst1=[]
        #check point notnull
        x=soup.find_all("div",class_="a-section table-padding")
        for i in x:
            new=i.find_all(class_="a-size-base prodDetAttrValue")
            for j in new:
                lst2.append(j.text)

            new1=i.find_all(class_="a-color-secondary a-size-base prodDetSectionEntry")
            for j in new1:
                lst1.append(j.text)

        dic = dict(zip(lst1,lst2))

        ASIN=dic[' ASIN ']
        Manufacturer = dic[' Manufacturer ']

        desc = soup.find_all("div",class_="a-section a-spacing-medium a-spacing-top-small")
        desc=desc[0].text

        final_lst.append([data[0],data[1],data[2],data[3],data[4],ASIN,Manufacturer,desc])
print(final_lst)
```

```python
#Converting list into dataframe
final_df = pd.DataFrame(final_lst, columns = ['Product_name','Product_price','Product_rating','No_of_reviews','Product_url','Product ASIN','Product Manufacturer','Product Description'])
```

```python
final_df.to_csv("final_dataset.csv",index=False)
```
